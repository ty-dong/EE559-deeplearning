```python
# Import Pytorch Package 
import torch
from torch import nn
from torch import optim
from torch.nn import functional as F
# Import Tool package
import dlc_practical_prologue as prologue
# Import Plot Package
import matplotlib.pyplot as plt
# Import the models and functions
from func import *
from models import *
# Other Package
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the data
N = 1000
train_input, train_target, train_classes, test_input, test_target, test_classes = prologue.generate_pair_sets(N)
```

# 1. Model Training

```python
# -------------------- Baseline Model -------------------- # 
# inpDim: Input dimension
inpDim = train_input.view(N, -1).size()[1]
# nb_hidden1, nb_hidden2: Hidden dimensions
nb_hidden1, nb_hidden2 = 160, 128
rounds = 15
mini_batch_size = 50
```

```python
print("Train Baseline Model:")
print("--------------------------------------------------")
model1, loss_round_list1, train_errors_list1, test_errors_list1 = PipelineBase(BaselineNet, rounds, mini_batch_size)
estiPerform(train_errors_list1, test_errors_list1)
print("The parameters number of Baseline Model:", sum(p.numel() for p in model1.parameters()))
```

```python
# -------------------- CNN Model -------------------- # 
print("Train CNN Model:")
print("--------------------------------------------------")
model2, loss_round_list2, train_errors_list2, test_errors_list2 = PipelineBase(CNN, rounds, mini_batch_size)
estiPerform(train_errors_list2, test_errors_list2)
print("The parameters number of Baseline Model:", sum(p.numel() for p in model2.parameters()))
```

```python
# -------------------- CNN Model with Auxiliary Loss -------------------- # 
print("CNN Model with Auxiliary Loss:")
print("--------------------------------------------------")
model3, loss_round_list3, train_errors_list3, test_errors_list3 = PipelineAux(CNNAuxiliaryLoss, rounds, mini_batch_size)
estiPerform(train_errors_list3, test_errors_list3)
print("The parameters number of Baseline Model:", sum(p.numel() for p in model3.parameters()))
```

# Model Comparison

```python
# Compare the models
seed = 0
plt.figure(figsize=(10, 5))
ax1 = plt.subplot(111)
ax1.plot(range(len(loss_round_list1[seed])), loss_round_list1[seed], label='Base Line')
ax1.plot(range(len(loss_round_list2[seed])), loss_round_list2[seed], label='Weight Sharing')
ax1.plot(range(len(loss_round_list2[seed])), loss_round_list3[seed][:len(loss_round_list2[seed])], label='Auxiliary Loss')
ax1.set_ylabel('Loss')
ax1.set_xlabel('Iteration')
ax1.set_title('Loss Convergence')
ax1.legend()
# plt.savefig('Loss-Convergence.png')

plt.figure(figsize=(10, 5))
ax2 = plt.subplot(111)
ax2.plot(range(len(train_errors_list1)), train_errors_list1, label='Base Line', marker='o')
ax2.plot(range(len(train_errors_list2)), train_errors_list2, label='Weight Sharing', marker='o')
ax2.plot(range(len(train_errors_list3)), train_errors_list3, label='Auxiliary Loss', marker='o')
ax2.set_ylabel('Train error rate')
ax2.set_xlabel('round')
ax2.set_title('Train Error Comparison')
ax2.legend()
# plt.savefig('Train-Error-Comparison')

plt.figure(figsize=(10, 5))
ax2 = plt.subplot(111)
ax2.plot(range(len(test_errors_list1)), test_errors_list1, label='Base Line', marker='o')
ax2.plot(range(len(test_errors_list2)), test_errors_list2, label='Weight Sharing', marker='o')
ax2.plot(range(len(test_errors_list3)), test_errors_list3, label='Auxiliary Loss', marker='o')
ax2.set_ylabel('Test error rate')
ax2.set_xlabel('round')
ax2.set_title('Test Error Comparison')
ax2.legend()
# plt.savefig('Test-Error-Comparison')

plt.show()
```

```python
import numpy as np
x = np.arange(6)
y_list = [train_errors_list1, test_errors_list1, train_errors_list2, test_errors_list2, train_errors_list3, test_errors_list3]
y = np.mean(np.array(y_list), axis=1)
e = np.std(np.array(y_list), axis=1)

plt.figure(figsize=(10,5))
plt.errorbar(x[1], y[1], e[1], linestyle='None', label ='Base Line:Test', marker='o') 
plt.errorbar(x[3], y[3], e[3], linestyle='None', label ='Weight Sharing:Test', marker='o')
plt.errorbar(x[5], y[5], e[5], linestyle='None', label ='Auxiliary Loss:Test', marker='o')
plt.xticks([])
plt.legend()
plt.title('Test error rate standard deviations')
# plt.savefig('Test-error-rate-standard-deviations')
```
